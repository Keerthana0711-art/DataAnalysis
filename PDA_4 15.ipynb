Encoding

*converting catogarical data into numarical data
label and one-Hot 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv(r"C:\Users\DELL\Desktop\pythonfiles\Salary_EDA.csv")
df.head()
```

```python
categorical_cols=['Education Level']
```

defeine and apply encoder

```python
encoder=OneHotEncoder(drop=None,sparse_output=False)
```

```python
encoded_data=encoder.fit_transform(df[categorical_cols])
```

```python
print (encoder_data)
```

encode is in the form of array.now we need to convert the encoded features 

convert the encoded into dataframes with catagoris as coloumn names

```python
encoded_df=pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(categorical_cols))
encoded_df.head()
```

```python
encoded_df.drop(columns=['Education Level_nan'])
```

```python
encoded_df.drop(columns=['Education Level_nan'],axis=1)
```

```python
encoded_df.drop(columns=['Education Level_nan'],axis=0)
```

```python
final_df=pd.concat([df, encoded_df],axis=1)
final_df.head()
```
